### fatorial calculation

```python
def factorial(n):                            #iterative
    product = 1
    while n > 0:
        product = product*n
        n = n-1
    return product
```

```python
factorial(5)
```

```python
def factorialr(n):                        #recursive
    if n == 0:
        return 1
    elif n == 1:
        return 1
    
    return n*factorialr(n-1)
```

```python
factorialr(5)
```

```python
import random as r
```

```python
r.randint(1,6)
```

```python
def dice(n):
    L = []
    for i in range(n):
        face = r.randint(1,6)
        L.append(face)
    return L
        
```

```python
dice(10)
```

```python
L1 = dice(10)
L2 = dice(10)

for d1,d2 in zip(L1,L2):
    print(d1,d2,d1+d2)
```

```python
def count(L1,L2,p,q):
    c=0
    for l1,l2 in zip(L1,L2):
        if l1 == p and l2 == q:
            c = c+1
    return c
```

```python
np.zeros((6,6), dtype = int)
```

```python
N = 1000
L1 = dice(N)
L2 = dice(N)
A = np.zeros((6,6), dtype = int)
for i in range(6):
    for j in range(6):
        A[i,j] = count(L1,L2,i+1,j+1)
```

```python
A
```

```python
import seaborn as sns
sns.set()
```

```python
sns.heatmap(A,annot = True)
```

### random Walk

```python
r.choice([-1,1])
```

```python
X = []
Y = []
N = 1000
x = 0
y = 0
for t in range(N):
    dx = r.choice([-1,1])
    dy = r.choice([-1,1])
    x = x + dx
    y = y + dy
    X.append(x)
    Y.append(y)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize = [10,10])
plt.plot(X,Y)
plt.show()
```

```python
def walk(N):
    X = []
    Y = []
    x = 0
    y = 0
    for t in range(N):
        dx = r.choice([-1,1])
        dy = r.choice([-1,1])
        x = x + dx
        y = y + dy
        X.append(x)
        Y.append(y)
    return X,Y
```

```python
N = 20000
X1,Y1 = walk(N)
X2,Y2 = walk(N)
X3,Y3 = walk(N)
plt.figure(figsize = [10,10])
plt.plot(X1,Y1)
plt.plot(X2,Y2)
plt.plot(X3,Y3)
plt.show()
```
